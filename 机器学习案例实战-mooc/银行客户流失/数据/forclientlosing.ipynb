```python
#去掉不相关的属性
```

```python
df.drop('EB',axis=1,inplace=True)
```

```python
dft=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
dft.columns
```

```python
#数据变换，编码
```

```python
for x in dft.columns:
    if df[x].dtype=='object':
        unique=len(df[x].unique())
        print(f'{x} has {unique} parts')
```

```python
df.isnull().sum()
```

```python
def DummiesDf(x,dummylist):
    for col in dummylist: 
        dummy=pd.get_dummies(x[col],prefix=col,dummy_na=False)
        x=x.drop(col,axis=1)
        x=pd.concat([x,dummy],axis=1,sort='False')
    return x
```

```python
def FindToDummy(df):
    tu=[]
    for col in df.columns:
        if df[col].dtype=='object':
            tu.append(col)
    return tu
            
```

```python
ls=FindToDummy(dft)
```

```python
ls
```

```python
dftdum=DummiesDf(dft,ls)
```

```python
dftdum.columns
```

```python
y=df['Exited']
```

```python
dftdumf=dftdum.drop('Exited',axis=1)
```

```python
dftdumf.columns
```

```python
#类不平衡时的处理
```

```python
dftdum['Exited'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
x_train,x_test,y_train,y_test=train_test_split(dftdumf,y,random_state=0)
```

```python
cls=tree.DecisionTreeClassifier()
```

```python
cls.fit(x_train,y_train)
```

```python
from sklearn.metrics import recall_score,roc_auc_score
```

```python
recall_score(y_test,cls.predict(x_test))
```

```python
roc_auc_score(y_test,cls.predict(x_test))
```
